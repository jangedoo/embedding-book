# Distance Metrics Demo

Compare cosine similarity, dot product, and Euclidean distance on vectors with different norms.

```python
import numpy as np

q = np.array([1.0, 0.0])
docs = {
    'same_direction_short': np.array([0.5, 0.0]),
    'same_direction_long': np.array([3.0, 0.0]),
    'slightly_angled': np.array([1.0, 0.4]),
    'orthogonal': np.array([0.0, 1.0]),
}

def cosine(a, b):
    return float(a @ b / (np.linalg.norm(a) * np.linalg.norm(b)))

for name, d in docs.items():
    print(name, {'dot': float(q @ d), 'cosine': cosine(q, d), 'euclidean': float(np.linalg.norm(q - d))})
```
