# Factorized Embeddings Demo

Compare parameter counts for full and factorized embedding layers.

```python
def full_params(V, d):
    return V * d

def factorized_params(V, r, d):
    return V * r + r * d

V, d = 250_000, 1024
for r in [16, 32, 64, 128, 256]:
    print(r, factorized_params(V, r, d), factorized_params(V, r, d) / full_params(V, d))
```
