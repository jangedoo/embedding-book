# Linear and ReLU Geometry Demo

Visualize how a linear projection and ReLU gate reshape a toy embedding space.

```python
import numpy as np

X = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]], dtype=float)
W = np.array([[1, 1], [1, -1], [-1, 0.5]], dtype=float)
b = np.array([0.0, 0.0, 0.2])

H_linear = X @ W.T + b
H_relu = np.maximum(0, H_linear)

print('linear activations')
print(H_linear)
print('relu activations')
print(H_relu)
```
